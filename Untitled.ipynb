```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
from sklearn.ensemble import IsolationForest
from scipy import stats
import pprint
```

```python
df = pd.read_excel("Superstore.xls")
df.head()
```

```python
df.info()
```

```python
def show_stats(lines, labels):
    
    # distribution
    for i, line in enumerate(lines):
        sns.distplot(line, label=labels[i])
    plt.title("Distribution")
    plt.ylabel("%")
    plt.legend()
    sns.despine()
    plt.show()
    
    # scatter distibution
    for i, line in enumerate(lines):
        plt.scatter(range(len(line)), np.sort(line), label=labels[i])
    plt.xlabel('index')
    plt.ylabel('value')
    plt.title('Distribution')
    plt.legend()
    sns.despine()
    plt.show()
    
    # describe
    print("Data Description")
    for i, line in enumerate(lines):
        print(labels[i])
        pprint.pprint(stats.describe(line))
    
```

```python
def detect_isolation_forest(df, line):
    isolation_forest = IsolationForest(n_estimators=100)
    isolation_forest.fit(df[line].values.reshape(-1,1))
    xx = np.linspace(df[line].min(), df[line].max(), len(df)).reshape(-1,1)
    anomaly_score = isolation_forest.decision_function(xx)
    outlier = isolation_forest.predict(xx)
    line_outlier = isolation_forest.predict(df[line].values.reshape(-1,1))
    plt.figure(figsize=(10,4))
    plt.plot(xx, anomaly_score, label='anomaly_score')
    plt.fill_between(xx.T[0], np.min(anomaly_score), np.max(anomaly_score), where=outlier==-1, color='r', alpha=.4, label='outlier region')
    plt.legend()
    plt.ylabel('anomaly score')
    plt.xlabel(line)
    plt.show()
    
    df['outlier isolation forest %s' %line]=line_outlier
    show_stats([df[df['outlier isolation forest %s' %line]==-1][line].values, df[df['outlier isolation forest %s' %line]==1][line].values],['outliers', 'inliers'])

```

```python
detect_isolation_forest(df, 'Profit')
```

```python
def detect_correlation(df, line1, line2):
    sns.regplot(x=line1, y=line2, data=df)
    sns.despine()
    plt.show()
```

```python
detect_correlation(df, "Sales", "Profit")
```
