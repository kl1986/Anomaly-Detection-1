```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_excel("Superstore.xls")
df.head()
```

```python
df.info()
```

```python
def show_stats(line):
        
    # distribution
    sns.distplot(df[line])
    plt.title("Sales Distribution")
    sns.despine()
    plt.show()
    
    # scatter distibution
    plt.scatter(range(df.shape[0]), np.sort(df[line].values))
    plt.xlabel('index')
    plt.ylabel(line)
    plt.title('%s Distribution' %line)
    sns.despine()
    plt.show()
    
    # scatter by date
    plt.scatter(df.sort_values(by='Order Date')['Order Date'], df.sort_values(by='Order Date')[line])
    plt.xlabel('Date')
    plt.ylabel(line)
    plt.title('%s by Date' %line)
    sns.despine()
    plt.show()
    
    # skewness and kurtosis
    print("Skewness and Kurtosis")
    print("Skewness: %f" % df[line].skew())
    print("Kurtosis: %f" % df[line].kurt())
    
    # describe
    print("Data Description")
    print(df[line].describe())
    
show_stats("Sales")
```

```python
def detect_isolation_forest(df, line):
    isolation_forest = IsolationForest(n_estimators=100)
    isolation_forest.fit(df[line].values.reshape(-1,1))
    xx = np.linspace(df[line].min(), df[line].max(), len(df)).reshape(-1,1)
    anomaly_score = isolation_forest.decision_function(xx)
    outlier = isolation_forest.predict(xx)
    line_outlier = isolation_forest.predict(df[line].values.reshape(-1,1))
    plt.figure(figsize=(10,4))
    plt.plot(xx, anomaly_score, label='anomaly_score')
    plt.fill_between(xx.T[0], np.min(anomaly_score), np.max(anomaly_score), where=outlier==-1, color='r', alpha=.4, label='outlier region')
    plt.legend()
    plt.ylabel('anomaly score')
    plt.xlabel(line)
    plt.show()
    print(sum(1 if x==-1 else 0 for x in line_outlier))
    print(len(line_outlier))
    
    return line_outlier
```

```python
line_outlier = detect_isolation_forest(df, 'Sales')
```
